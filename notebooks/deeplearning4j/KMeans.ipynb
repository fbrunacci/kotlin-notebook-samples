```kotlin
import krangl.DataFrame
import krangl.readDelim
import org.apache.commons.csv.CSVFormat
import org.deeplearning4j.clustering.algorithm.Distance
import org.deeplearning4j.clustering.cluster.Point.toPoints
import org.deeplearning4j.clustering.kmeans.KMeansClustering
import org.nd4j.linalg.factory.Nd4j
import java.awt.Point
import java.io.StringReader
import org.deeplearning4j.clustering.cluster.Cluster
import org.nd4j.linalg.util.ArrayUtil

val sixMap = sample.rows.map { row -> doubleArrayOf(row["X1"] as Double, row["X2"] as Double) }.toTypedArray()
val sixINDArray: INDArray = Nd4j.create(ArrayUtil.flattenDoubleArray(sixMap), intArrayOf(sixMap.size, 2))
```

```kotlin
import org.apache.commons.lang3.time.StopWatch
val sw = StopWatch()

data class Point(val x:Float, val y:Float, val label:String)
val points = mutableListOf<Point>()

val kMeansClustering = KMeansClustering.setup(6, 5, Distance.EUCLIDEAN, true)
println("Start Clustering " + sixMap.size + " points")
sw.reset()
sw.start()
val clusterSet = kMeansClustering.applyTo(toPoints(sixINDArray))
sw.stop();
println("Time taken to run clustering ${sw.time} ms"  );
```

```kotlin
val clsterLst: List<Cluster> = clusterSet.getClusters()
var i = 0
for( cluster in clsterLst ) {
    i++
    for(point in cluster.points) {
        val coord = point.array.data().asFloat()
        points.add(Point(coord[0],coord[1],"cluster$i"))
    }
}
val pointsDf = points.asDataFrame()

// TODO limit nb points for training and after launch classifyPoint...
//val pointClassification = clusterSet.classifyPoint(points[0])
//println(pointClassification)
```
